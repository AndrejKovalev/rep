Выбросы для категориальных данных.
Делаем оценку экспертно.
Хирургия, принимает только два значения, выбросов нет.
Температура конечностей от 1 до 4, выбросов нет.
Боль, уровни от 1 до 5, выбросов нет.

```python
q1_p = df['pulse'].quantile(0.25)      # выбросы по пульсу
q3_p = df['pulse'].quantile(0.75)
iqr_p = q3_p - q1_p
lower_bound_p = q1_p - (1.5 * iqr_p)
upper_bound_p = q3_p + (1.5 * iqr_p)
remove_outliers_p = df[~df['pulse'].between(lower_bound_p, upper_bound_p, inclusive=True)]
remove_outliers_p.head()
```

```python
q1_r = df['respiratory rate'].quantile(0.25)      # выбросы по частоте дыхания
q3_r = df['respiratory rate'].quantile(0.75)
iqr_r = q3_r - q1_r
lower_bound_r = q1_r - (1.5 * iqr_r) 
upper_bound_r = q3_r + (1.5 * iqr_r)
remove_outliers_r = df[~df['respiratory rate'].between(lower_bound_r, upper_bound_r, inclusive=True)]
remove_outliers_r.head()
```

```python
q1_t = df['rectal temperature'].quantile(0.25)      # выбросы по температуре
q3_t = df['rectal temperature'].quantile(0.75)
iqr_t = q3_t - q1_t
lower_bound_t = q1_t - (1.5 * iqr_t) 
upper_bound_t = q3_t + (1.5 * iqr_t)
remove_outliers_t = df[~df['rectal temperature'].between(lower_bound_t, upper_bound_t, inclusive=True)]
remove_outliers_t.head()
```

Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
df.info()
```

```python
(df.isna().mean() * 100).round(2) # % пропусков по каждому столбцу
```

```python
(df.isna().mean() * len(df)) # количество пропусков по каждому столбцу
```

```python
df_unhealthy=df[((df['pain'] >= 4) & (df['surgery?'] == 1)) | 
                ((df['pain'] >= 4) & (df['outcome'] == 3)) | 
                ((df['pain'] >= 4) & (df['rectal temperature'] >= 39)) | 
                ((df['outcome'] == 3) & (df['rectal temperature'] >= 39)) | 
                ((df['outcome'] == 3) & (df['surgery?'] == 1))]
df_unhealthy.info()
(df_unhealthy.isna().mean() * len(df_unhealthy)).round(0)
# разделим датафрейм на больных и здоровых лошадей, считаем если лошадь испытывает сильную боль и была операция
# либо лошадь усыпили либо высокая температура, значит лошадь больна вне завимости от возраста.
```

```python
df_healthy=df[~(((df['pain'] >= 4) & (df['surgery?'] == 1)) | 
                ((df['pain'] >= 4) & (df['outcome'] == 3)) | 
                ((df['pain'] >= 4) & (df['rectal temperature'] >= 39)) | 
                ((df['outcome'] == 3) & (df['rectal temperature'] >= 39)) | 
                ((df['outcome'] == 3) & (df['surgery?'] == 1)))]
df_healthy.info()
(df_healthy.isna().mean() * len(df_healthy)).round(0)
# остальных лошадей считаем здоровыми
```

```python
print(df_unhealthy.groupby('Age')['rectal temperature'].median())
df_unhealthy_age = df_unhealthy.copy()
df_unhealthy_age['rectal temperature'] = df_unhealthy_age['rectal temperature']\
.fillna(df_unhealthy.groupby('Age')['rectal temperature'].transform('median'))
# жеребята всегда теплее взрослых лошадей (датафрейм также это подтверждает), 
# поэтому температуру будем определять отдельно для этих групп. 
print(df_unhealthy.groupby('Age')['pulse'].median())
df_unhealthy_age['pulse'] = df_unhealthy_age['pulse'].fillna(df_unhealthy.groupby('Age')['pulse'].transform('median'))
# пульс жеребят также сильно отличается, он выше чем у взрослых лошадей
print(df_unhealthy.groupby('Age')['respiratory rate'].median())
df_unhealthy_age['respiratory rate'] = df_unhealthy_age['respiratory rate']\
.fillna(df_unhealthy.groupby('Age')['respiratory rate'].transform('median'))
# частота дыхания жеребят также  отличается, он выше чем у взрослых лошадей
df_unhealthy_age['temperature of extremities'] = df_unhealthy_age['temperature of extremities']\
.fillna(df_unhealthy.groupby('Age')['temperature of extremities'].transform('median'))
# частота дыхания жеребят также  отличается, он выше чем у взрослых лошадей
df_unhealthy_age['pain'] = df_unhealthy_age['pain']\
.fillna(df_unhealthy.groupby('Age')['pain'].transform(lambda x: x.mode()[0]))
# боль заменяем модой
df_unhealthy_age.info()
```

```python
print(df_unhealthy['pain'].mode()[0])
print(df_unhealthy_age['pain'].mode()[0])
# проверим, что модой заполненеи прошло корерктно, значения равны, соответственно все ОК
```
